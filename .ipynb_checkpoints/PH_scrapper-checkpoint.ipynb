```python
from datetime import datetime, timedelta
import pandas as pd    

def from_iso(iso_str):
    return datetime.strptime(iso_str, "%Y-%m-%d").date()

def crawl_one_day(date, after_cursor=""):
    posted_after = date.isoformat()
    posted_before = (date + timedelta(days=1)).isoformat()

    # print(posted_before, posted_after, after_cursor)
    variables = {"postedBefore": posted_before, "postedAfter": posted_after, "after": after_cursor}

    result = client.execute(query, variable_values=variables)

    if(result["posts"]["pageInfo"]["hasNextPage"]):
        next_nodes = crawl_one_day(date, result["posts"]["pageInfo"]["endCursor"])
        return result["posts"]["nodes"] + next_nodes
    
    return result["posts"]["nodes"] 

def append_result(result):
    df = pd.json_normalize(result)
    df.drop_duplicates(subset='id', keep="first", inplace=True)
    df.to_csv('ProductHuntProducts.csv', mode='a', index=False, header=False)
```

```python
day_one = crawl_one_day(from_iso("2024-07-09"))
```

```python
from tqdm import tqdm
import pandas

# pandas.date_range(from_iso("2024-06-08"),from_iso("2024-07-08")-timedelta(days=1),freq='d')
sdate = from_iso("2024-04-15")
edate = from_iso("2024-05-15")
date_list = [sdate+timedelta(days=x) for x in range((edate-sdate).days)]
date_list.reverse()
# date_list

for day in tqdm(date_list):
    append_result(crawl_one_day(day))
```
